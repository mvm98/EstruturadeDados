```python
class ListaSequencial:
  #Definindo função ListaSequencial
  def __init__(self,capacidade):
    self.capacidade = capacidade #Definindo a capacidade a partir do valor passado na criação da Lista
    self.valores = [0]*capacidade #Calculo da quantidade de valores que a lista poderá comportar
    self.tamanho = 0 #Tamanho inicial
  #Definindo função para saber se a lista é vazia
  def vazia(self):
    return self.tamanho == 0
  #Definindo função para saber se a lista é cheia
  def cheia(self):
    return self.tamanho == self.capacidade
  #Definindo função para descobrir o tamanho da lista
  def tamanho(self):
    return self.tamanho
  #Função para saber o tamanho da lista
  def obter_valor(self, posicao):
    if posicao >= 0 or posicao <= self.tamanho: #Condição if para retornar o valor n(posição) da lista
      return self.valores[posicao]
    else:
      print('Posição inválida')
  #Função para modificar um valor da lista
  def modificar_valor(self, posicao, valor):
    if 0 <= posicao < self.tamanho: #Sendo 'posicao' uma posição válida na lista
      self.valores[posicao] = valor #Troca o valor da 'posicao' por 'valor' 
    else:
      print('Posição inválida')
  #Função para inserir valor na lista
  def inserir(self, posicao, valor):
    if self.cheia(): #Inicialmente checamos se a lista está cheia
      print('Lista cheia')
    if posicao < 0 or posicao > self.tamanho:
      print('Posição inválida') #Depois checamos se a posição em que queremos inserir um valor é válida
    for i in range(self.tamanho-1, posicao-1, -1): #Sendo a posição válida temos um for que inicia-se no final da lista se 'deslocando' até a posição imediatamente anterior à que desejo inserir um valor
      self.valores[i+1] = self.valores[i] #Troco o valor 'posterior' pelo valor que se encontra imediatamente antes dele
    self.valores[posicao] = valor #Insiro o 'valor' na posição 'posicao' desejada
    self.tamanho += 1 #Incremento o tamanho da lista
  #Função para retirar um valor da lista
  def retirar(self, posicao):
    if self.vazia():
      print('Lista vazia')
    if posicao < 0 or posicao >= self.tamanho:
      print('Posição inválida')
    valor = self.valores[posicao] #Identifico o valor para poder retorná-lo depois
    for i in range(posicao+1, self.tamanho): #Com essa linha inicio o for no valor imediatamente depois do escolhido para ser retirado, até o final da lista
      self.valores[i-1] = self.valores[i] #Troco cada um dos valores 'anteriores' pelo valor que se encontra imediatamente depois dele
    self.tamanho -= 1 #Altero o tamanho da lista para representar essa retirada
    return valor #Retorno o valor retirado
```

```python
def teste_lista_sequencial():
  
  #Teste criação da lista vazia
  lista = ListaSequencial(5)
  if lista.vazia() == True and lista.cheia() == False and lista.obter_tamanho() == 0:
      print("Teste criação da lista vazia: OK")
  else:
      print("Teste criação da lista vazia: FALHOU")
  
  #Teste obter/alterar valor em posição válida
  lista.inserir(0, 1)
  lista.modificar_valor(0, 2)
  if lista.obter_valor(0) == 2:
      print("Teste obter/alterar valor em posição válida: OK")
  else:
      print("Teste obter/alterar valor em posição válida: FALHOU")
  
  #Teste obter/alterar valor em posição inválida
  valor_obtido = lista.obter_valor(1)
  if valor_obtido == None:
      print("Teste obter valor em posição inválida: OK")
  else:
      print("Teste obter valor em posição inválida: FALHOU")
  
  try:
      lista.modificar_valor(1, 3)
      print("Teste alterar valor em posição inválida: FALHOU")
  except IndexError:
      print("Teste alterar valor em posição inválida: OK")
  
  #Teste inserir elemento em posição válida
  lista.inserir(0, 3)
  if lista.obter_tamanho() == 2 and lista.obter_valor(0) == 3 and lista.obter_valor(1) == 2:
      print("Teste inserir elemento em posição válida: OK")
  else:
      print("Teste inserir elemento em posição válida: FALHOU")
  
  #Teste inserir elemento em posição inválida
  try:
      lista.inserir(3, 4)
      print("Teste inserir elemento em posição inválida: FALHOU")
  except IndexError:
      print("Teste inserir elemento em posição inválida: OK")
  
  #Teste retirar elemento em posição válida
  valor_retirado = lista.retirar(1)
  if valor_retirado == 2 and lista.obter_tamanho() == 1 and lista.obter_valor(0) == 3:
      print("Teste retirar elemento em posição válida: OK")
  else:
      print("Teste retirar elemento em posição válida: FALHOU")
  
  #Teste retirar elemento em posição inválida
  valor_retirado = lista.retirar(1)
  if valor_retirado == None:
      print("Teste retirar elemento em posição inválida: OK")
  else:
      print("Teste retirar elemento em posição inválida: FALHOU")
```
